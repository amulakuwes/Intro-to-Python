```python
#build a simple calculator application that does
#addition, subtraction, multiplication, division
a = int(input("Enter a number a"))
b = int(input("Enter a number b"))

#addition
is_result = a + b
print(f"Addition: {a} + {b} = {is_result}")
#subtraction
is_result = a - b
print(f"Subtraction: {a} - {b} = {is_result}")
#multiplication
is_result = a * b
print(f"Multiplication: {a} * {b} = {is_result}")
#division
is_result = a / b
print(f"Division: {a} / {b} = {is_result}")

```
