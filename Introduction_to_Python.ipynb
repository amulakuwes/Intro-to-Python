```python
#Create a grading system for mathematics
score = int(input("Enter your score"))
#80 - A
#70 - A-
#60 - B+
#50 - B
#40 - C+

if score > 80:
    print("The grade is an A")
elif score >= 70:
    print("The grade is an A-")
elif score >= 60:
    print("The grade is a B+")
elif score >= 50:
    print("The grade is a B")
elif score >= 40:
    print("The grade is a C+")
else:
    print("The grade is a fail")
```

```python
#Create a generation calculator
#baby boomer, milleial, gen Alpha, gen Z
age = int(input("Enter your age"))

if age <= 10:
  print("This is Gen Alpha")
elif age <= 27:
  print("This is Gen Z")
elif age <= 40:
  print("This is Millenial")
elif age <= 66:
  print("This is baby boomer")
else:
  print("This is undefined")
```

```python
#Create a 2027 election vetting system
#age < 18 - they should not be allowed to vote
#age > 18 and they should have a national ID

age = int(input("Enter your age"))
national_id = str(input("Do you have a national ID?"))

if age < 18:
    print("you are not allowed to vote")
elif age >= 18 and national_id == ("yes"):
    print("You are allowed to vote")
else:
    print("You are not allowed to vote")
```

```python
#assignment
#create a fruit tuple
#perform indexing, slicing, looping

#indexing in tuple

fruit = ("Pineapple", "Strawberry", "Grapes", "Avocado", "Blueberry")
print(fruit)

print(fruit[0])
print(fruit[1])
print(fruit[2])
print(fruit[3])
print(fruit[4])

print(fruit[-1])
print(fruit[-2])
```

```python
#slicing in tuple

fruit = ("Pineapple", "Strawberry", "Grapes", "Avocado", "Blueberry", "Raspberry")

print(fruit[:])
print(fruit[:3])
print(fruit[2:])
print(fruit[1:4])
```

```python
#looping in tuple

fruit = ("Pineapple", "Strawberry", "Grapes", "Avocado", "Blueberry", "Raspberry")
for fruit in fruit:
  print(fruit)
```

```python
fruit = ("Pineapple", "Strawberry", "Grapes", "Avocado", "Blueberry", "Raspberry")
for index, fruit in enumerate (fruit):
  print(f"Index {index}:{fruit}")
```
